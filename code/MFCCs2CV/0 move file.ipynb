```python
label_path =  './track3_R1_label.txt'
print(label_path)
```

```python
with open(label_path,'r') as f:
    train_labels = f.readlines()
train_labels
```

```python
name_list = []
label_list =[] 
for item in train_labels:
    item = item.strip()
    item = item.split(' ')
    name_list.append(item[0])
    label_list.append(item[1])
print(name_list[0:10], label_list[0:10])
```

```python
data_path = ori_path
ori_file_list = os.listdir(data_path)
# print(ori_file_list)
# for i in range(len(ori_file_list)):
#     print(i)
#     if i > 100:
#         break
```

```python
# tgt_fake_path
# tgt_genuine_path 

from tqdm import tqdm
for i in tqdm(range(len(ori_file_list))):
    file_name = ori_file_list[i]
    index = name_list.index(file_name)
#     print(index)
    label = label_list[index]
    os.rename(os.path.join(data_path, file_name) ,'./Test/'+label+'/'+file_name)
```
